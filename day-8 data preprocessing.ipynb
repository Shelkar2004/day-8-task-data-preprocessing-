```python
#Fill NULL values
#Fill values with Mean

#identify numeric columns

numeric_columns = dataset.select_dtypes(include=np.number).columns.tolist()
numeric_columns
```

```python
#Filling values with Mean

dataset[numeric_columns] = dataset[numeric_columns].fillna(dataset[numeric_columns].mean())
```

```python
#Fill values with Mode (Most Frequent)

#identify categorical columns

categorical_columns = ['Gender', 'Salary','Expenditure']
categorical_columns
```

```python
##Filling values with Mode

dataset[categorical_columns] = dataset[categorical_columns].fillna(dataset[categorical_columns].mode().iloc[0])
```

```python
#Drop non required rows

#Drop rows for NaN values in 'Smoking'

dataset.dropna(subset=['Savings Percentage'], axis=0, inplace= True)
dataset
```

```python
from sklearn import svm

# Sample data
X = np.array([[35, 81.25],
              [35, 82.86],
              [36, 92.31],
              [42, 80.65],
              [29, 91.94]])
y = np.array([0, 0, 1, 0, 1]) 
# Train an SVM classifier
clf = svm.SVC(kernel='linear')
clf.fit(X, y)

# Plot the data points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, marker='o', edgecolors='k')

# Plot the decision boundary (hyperplane)
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Create grid to evaluate model
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50),
                     np.linspace(ylim[0], ylim[1], 50))

# Decision function to get the hyperplane
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundary and margins
ax.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])

# Highlight support vectors
ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Expenditure Percentage')
plt.title('Hyperplane Plot for SVM Classifier')

# Display the plot
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

print("X_train: ", X_train.shape)
print("X_test: ", X_test.shape)
print("y_train: ", y_train.shape)
print("y_test: ", y_test.shape)
df
```
